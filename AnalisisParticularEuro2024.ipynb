### Es necesario convertir string a diccionarios

```python
import ast

def convert_to_list(location_str):
  try:
    return ast.literal_eval(location_str)
  except (SyntaxError, ValueError):
    return None

df['location_y'] = df['location_y'].apply(convert_to_list)
```

## Asignar tipo de possession a las secuencias

```python
def assign_in_possession_phase(event):
    if event['play_pattern'] == 'From Counter':
        return 'Counterattack'
    # If 'under_pressure' is True, phase is 'Consolidate'
    # elif event['under_pressure']:
    #     return 'Consolidate'
    else:
        x = event['location_y'][0]
        print(x)
        if x <= 40:
            return 'Buildup'
        elif x <= 80:
            return 'Progression'
        else:
            return 'Finishing'

def assign_out_of_possession_phase(event, opponent_players):
    if event['in_possession_phase'] == 'Counterattack':
        return 'Regroup'
    elif event['counterpress']:
        return 'Counterpress'
    else:
        opponent_players['x'] = opponent_players['location_x'].apply(lambda loc: loc[0])

        opponent_players['x_adj'] = 120 - opponent_players['x']

        attacking_third = opponent_players[(opponent_players['x_adj'] >= 80) & (opponent_players['x_adj'] <= 120)].shape[0]
        middle_third = opponent_players[(opponent_players['x_adj'] > 40) & (opponent_players['x_adj'] < 80)].shape[0]
        defensive_third = opponent_players[(opponent_players['x_adj'] >= 0) & (opponent_players['x_adj'] <= 40)].shape[0]

        if attacking_third >= 4:
            return 'High Block'
        elif defensive_third >= 6:
            if event['under_pressure']:
                return 'Low Press'
            else:
                return 'Low Block'
        elif middle_third >= 4:
            if event['under_pressure']:
                return 'Mid Press'
            else:
                return 'Mid Block'
        else:
            return 'Unidentified'

def assign_set_piece_phase(event, set_piece_counts):
    set_piece_types = ['From Corner', 'From Free Kick', 'From Throw In']
    if event['play_pattern'] in set_piece_types:
        count = set_piece_counts.get(event['possession'], 0)
        if count < 5:
            set_piece_counts[event['possession']] = count + 1
            return True
    return False

set_piece_counts = {}

in_possession_phases = []
out_of_possession_phases = []
event_ids = []

grouped = df.groupby('event_id')

for event_id, group in grouped:
    event_data = group.iloc[0]
    on_ball_team = event_data['team']
    home_team = event_data['home_team']
    away_team = event_data['away_team']

    opponent_team = away_team if on_ball_team == home_team else home_team
    print(on_ball_team, opponent_team)

    opponent_players = group[group['teammate'] == False].copy()

    print(len(opponent_players))
    event = {
        'play_pattern': event_data['play_pattern'],
        'under_pressure': event_data['under_pressure'],
        'location_y': event_data['location_y'],
        'in_possession_phase': None,
        'counterpress': event_data.get('counterpress', False),
        'possession': event_data['possession']
    }

    if assign_set_piece_phase(event_data, set_piece_counts):
        in_phase = 'Set Piece For'
        out_phase = 'Set Piece Against'
    else:
        in_phase = assign_in_possession_phase(event)
        event['in_possession_phase'] = in_phase
        out_phase = assign_out_of_possession_phase(event, opponent_players)
    df.loc[group.index, 'in_possession_phase'] = in_phase
    df.loc[group.index, 'out_of_possession_phase'] = out_phase
    event_ids.append(event_id)

df['timestamp'] = pd.to_datetime(df['timestamp'])

df = df.sort_values('timestamp').reset_index(drop=True)

unknown_mask = df['in_possession_phase'].isin(['Unidentified', 'MissingSB', None])
known_mask = ~unknown_mask

phase_df = df[['timestamp', 'event_id', 'in_possession_phase', 'out_of_possession_phase']].drop_duplicates('event_id')

phase_df['ffill_in_phase'] = phase_df['in_possession_phase'].fillna(method='ffill')
phase_df['ffill_out_phase'] = phase_df['out_of_possession_phase'].fillna(method='ffill')
phase_df['ffill_time'] = phase_df['timestamp'].fillna(method='ffill')

phase_df['bfill_in_phase'] = phase_df['in_possession_phase'].fillna(method='bfill')
phase_df['bfill_out_phase'] = phase_df['out_of_possession_phase'].fillna(method='bfill')
phase_df['bfill_time'] = phase_df['timestamp'].fillna(method='bfill')

phase_df['ffill_time_diff'] = (phase_df['timestamp'] - phase_df['ffill_time']).dt.total_seconds().abs()
phase_df['bfill_time_diff'] = (phase_df['bfill_time'] - phase_df['timestamp']).dt.total_seconds().abs()

def fill_unknown_phases(row):
    if pd.isna(row['in_possession_phase']) or row['in_possession_phase'] in ['Unidentified', 'MissingSB', None]:
        if row['ffill_time_diff'] <= 3:
            return row['ffill_in_phase'], row['ffill_out_phase']
        elif row['bfill_time_diff'] <= 3:
            return row['bfill_in_phase'], row['bfill_out_phase']
        else:
            return row['in_possession_phase'], row['out_of_possession_phase']
    else:
        return row['in_possession_phase'], row['out_of_possession_phase']

phase_df[['in_possession_phase_filled', 'out_of_possession_phase_filled']] = phase_df.apply(
    fill_unknown_phases, axis=1, result_type='expand'
)

phase_mapping = phase_df.set_index('event_id')[['in_possession_phase_filled', 'out_of_possession_phase_filled']]

df = df.merge(phase_mapping, how='left', left_on='event_id', right_index=True)

df['in_possession_phase'] = df['in_possession_phase_filled']
df['out_of_possession_phase'] = df['out_of_possession_phase_filled']

df.drop(['in_possession_phase_filled', 'out_of_possession_phase_filled'], axis=1, inplace=True)
```

## Obtener porcentajes para cada equipo

```python
import pandas as pd
import numpy as np

required_columns = [
    'event_id', 'match_id_x', 'team', 'home_team', 'away_team', 'possession_team',
    'in_possession_phase', 'out_of_possession_phase'
]
for col in required_columns:
    if col not in df.columns:
        raise ValueError(f"Required column '{col}' is missing in the DataFrame.")

unique_events = df.groupby('event_id').first().reset_index()

team_phases = []

for idx, event in unique_events.iterrows():
    possession_team = event['team']

    home_team = event['home_team']
    away_team = event['away_team']
    teams = [home_team, away_team]
    non_possession_team = [team for team in teams if team != possession_team][0]

    in_phase = event['in_possession_phase']
    out_phase = event['out_of_possession_phase']

    team_phases.append({
        'event_id': event['event_id'],
        'team': possession_team,
        'match_id': event['match_id_x'],
        'phase': in_phase,
        'possession_status': 'in_possession'
    })

    team_phases.append({
        'event_id': event['event_id'],
        'team': non_possession_team,
        'match_id': event['match_id_x'],
        'phase': out_phase,
        'possession_status': 'out_of_possession'
    })

team_phases_df = pd.DataFrame(team_phases)

for index, row in unique_goal_events.iterrows():
    match_id = row['match_id']
    home_team = row['home_team']
    away_team = row['away_team']
    goal_minute = row['goal_minute']
    team_scored = row['team_scored']

    print(f"Match ID: {match_id}, Home Team: {home_team}, Away Team: {away_team}, Goal Minute: {goal_minute}, Team Scored: {team_scored}")

    for team in [home_team, away_team]:
      print(f"\n=== Phase Percentages for Team: {team} and match {match_id}===")

      team_data = team_phases_df[(team_phases_df['team'] == team) & (team_phases_df['match_id'] == match_id)]

      total_in_possession = team_data[team_data['possession_status'] == 'in_possession'].shape[0]
      total_out_of_possession = team_data[team_data['possession_status'] == 'out_of_possession'].shape[0]
      total_events = total_in_possession + total_out_of_possession

      possession_percentage = (total_in_possession / total_events) * 100 if total_events > 0 else 0
      out_of_possession_percentage = (total_out_of_possession / total_events) * 100 if total_events > 0 else 0

      print(f"\nPossession Percentage: {possession_percentage:.2f}%")
      print(f"Out-of-Possession Percentage: {out_of_possession_percentage:.2f}%")

      in_possession_data = team_data[team_data['possession_status'] == 'in_possession']
      out_of_possession_data = team_data[team_data['possession_status'] == 'out_of_possession']

      known_in_phases = in_possession_data[~in_possession_data['phase'].isin(['Unidentified', 'MissingSB', None])]
      known_out_phases = out_of_possession_data[~out_of_possession_data['phase'].isin(['Unidentified', 'MissingSB', None])]

      in_phase_counts = known_in_phases['phase'].value_counts()
      in_phase_percentages = (in_phase_counts / total_in_possession) * 100 if total_in_possession > 0 else 0

      out_phase_counts = known_out_phases['phase'].value_counts()
      out_phase_percentages = (out_phase_counts / total_out_of_possession) * 100 if total_out_of_possession > 0 else 0

      print("\nIn-Possession Phase Percentages:")
      if isinstance(in_phase_percentages, pd.Series):
          print(in_phase_percentages.to_string())
      else:
          print("No data for in possession phases.")

      print("\nOut-of-Possession Phase Percentages:")
      if isinstance(out_phase_percentages, pd.Series):
          print(out_phase_percentages.to_string())
      else:
          print("No data for out of possession phases.")

      unknown_in_phases = total_in_possession - known_in_phases.shape[0]
      unknown_in_percentage = (unknown_in_phases / total_in_possession) * 100 if total_in_possession > 0 else 0
      print(f"\nPercentage of Unknown In-Possession Phases: {unknown_in_percentage:.2f}%")

      unknown_out_phases = total_out_of_possession - known_out_phases.shape[0]
      unknown_out_percentage = (unknown_out_phases / total_out_of_possession) * 100 if total_out_of_possession > 0 else 0
      print(f"Percentage of Unknown Out-of-Possession Phases: {unknown_out_percentage:.2f}%")
```
